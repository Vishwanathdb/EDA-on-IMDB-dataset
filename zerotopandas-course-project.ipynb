## Asking and Answering Questions


#### Q1: `Top 20 Actors` acted in more number of movies form `2006 to 2016` ?

To find the actors having more demand and acted in more number of movies from 2006 to 2016.

````python
survey_df.Actors
````

````python
def split_multicolumn(column) :
    result_df = column.to_frame()
    options = []
    
    for ind, value in column.items() :
        
        for option in value.split(',') :
            
            if not option in result_df :
                options.append(option)
                result_df[option] = False
            
            result_df.at[ind, option] = True
    
    return result_df[options]
````

````python
all_actors_df = split_multicolumn(survey_df.Actors)
all_actors_df
````

````python
top_twenty_actors = all_actors_df.sum().sort_values(ascending=False).head(20)

top_twenty_actors
````

````python
plt.figure(figsize=(12,12))

plt.title('Top 20 Actors acted in most number of movies')
plt.xlabel('Number Of Movies')
plt.ylabel('Actors')
sns.barplot(x=top_twenty_actors, y=top_twenty_actors.keys());
````

#### Q2: Which are the `Top 10` movies according to `Rating` ?

Information about Top 10 movies according to Rating.

````python
survey_df.sort_values('Rating', ascending=False).head(10)
````

#### Q3: Which is the most watched `Gener` movies `2006 to 2016`?

To get information about the popular Genre.

````python
all_genre = split_multicolumn(survey_df.Genre)
all_genre
````

````python
total_count_genre = all_genre.sum().sort_values(ascending=False)
total_count_genre
````

````python
plt.figure(figsize=(16,16))

plt.title('Most Watched Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')

sns.barplot(x=total_count_genre, y=total_count_genre.keys());
````

#### Q4: Classify movies based on `Rating greater than 7 and less than 7` on` Year V/S Runtime Graph`. 

To understand people willingness to watch long content year by year and their rating variation.

````python
rating_df = survey_df['Rating'].to_frame()
rating_df['Runtime'] = survey_df['Runtime (Minutes)']
rating_df['Year'] = survey_df['Year']
rating_df
````

````python
rating_df['Greater than 7'] = False
for i in survey_df.Rank:
    if rating_df.at[i-1, 'Rating'] >= 7 :
        rating_df.at[i-1, 'Greater than 7'] = True

rating_df
````

````python
sns.scatterplot(x='Year', y='Runtime', hue='Greater than 7', data=rating_df )

plt.title('Rating on Year V/S RunTime')
plt.xlabel('Year')
plt.ylabel('RunTime');
````

#### Q5: What is the effect of `Votes` on `Revenue generated` for `Top 10 Movies based on Rating` ?

To understand the people willingness on voting and actually generating revenue to movie.

````python
top_ten_rated_movies = survey_df.sort_values('Rating', ascending=False).head(10)
top_ten_rated_movies
````

````python
plt.figure(figsize=(12,8))

plt.title('Votes v/s Revenue')
plt.xlabel('Votes')
plt.ylabel('Revenue')

sns.lineplot(x=top_ten_rated_movies.Votes, y=top_ten_rated_movies['Revenue (Millions)']);
````
